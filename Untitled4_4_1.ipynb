```python
import requests
from pathlib import Path
from tqdm import tqdm  # Optional, for progress bar

def download_file(url, save_path, chunk_size=128, show_progress=True):
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    progress_bar = tqdm(total=total_size, unit='B', unit_scale=True, disable=not show_progress)

    with open(save_path, 'wb') as file, progress_bar:
        for data in response.iter_content(chunk_size=chunk_size):
            progress_bar.update(len(data))
            file.write(data)

    return save_path

# URL of the file to download
urlVlt = "https://alice-open.oss-cn-zhangjiakou.aliyuncs.com/mPLUG/ViT-L-14.tar"
urlDataset = "https://www.robots.ox.ac.uk/~maxbain/frozen-in-time/data/MSRVTT.zip"
urlPre = "http://tjfd2.oss-cn-zhangjiakou.aliyuncs.com/mplug2/mPLUG2_Pretrain.pth"
save_path = Path("MPlug2DataSet.zip")
download_file(urlDataset, save_path)
save_path = Path("ViT-L-14.pth")
download_file(urlVlt, save_path)
save_path = Path("pretrained_mplug_2.pth")
download_file(urlPre, save_path)



!unzip MPlug2DataSet.zip
```

```python
train_path = '/content/MSRVTT/videos/train_list_new.txt'
test_path = '/content/MSRVTT/videos/test_list_new.txt'


with open(train_path, 'r') as file:
  train_array = [line.strip() for line in file]
  print("Train loaded successfully.")

with open(test_path, 'r') as file:
  test_array = [line.strip() for line in file]
  print("Test loaded successfully.")

```

```python
import copy

train_images = []
train_annotations = []
test_images = []
test_annotations = []
```

```python
i = 0
for image in data_array['images']:
  if image['id'] in train_array:
    train_images.append(image)
  else:
    test_images.append(image)

print(len(data_array['images']), len(train_images), len(test_images), len(train_images) + len(test_images))
```

```python
i = 0
for annotation in data_array['annotations']:
  if annotation['image_id'] in train_array:
    train_annotations.append(annotation)
  else:
    test_annotations.append(annotation)

print(len(data_array['annotations']), len(train_annotations), len(test_annotations), len(train_annotations) + len(test_annotations))
```

```python
train = copy.deepcopy(data_array)
test = copy.deepcopy(data_array)

train['images'] = train_images
test['images'] = test_images

train['annotations'] = train_annotations
test['annotations'] = test_annotations

```

```python


train_path = '/content/train.json'
test_path = '/content/test.json'

with open(train_path, 'w') as json_file:
    json.dump(train, json_file, indent=4)

with open(test_path, 'w') as json_file:
    json.dump(test, json_file, indent=4)
```
